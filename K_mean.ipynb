```python
import numpy as np
import pandas as pd
from sklearn import neighbors
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Tạo dữ liệu
X= -2 * np.random.rand(100,2) # np.random.rand(d1,d2,...,dn) : tạo vecto d_1 dimention, mỗi dimention gồm d2 vecto, ... , mỗi dimention gồm d_n mảng (note: giá trị random trong nữa khoảng [0,1))
X1 = 1 + 2 * np.random.rand(50,2)
X[50:100, :] = X1 # từ phần tử thứ 50 tới 99 gán cho vecto X1

# visualization
plt.scatter(X[ : , 0], X[ :, 1], s = 50, c = 'r',marker = '^')  
plt.show()
```

```python
Kmean = KMeans(n_clusters=2) # hai clusterring
Kmean.fit(X)
```

```python
Centroid = Kmean.cluster_centers_ # return tọa độ của enter

Centroid

```

```python
# Visualization
x_lin= np.arange(-3, 3.3, 0.01) # Tạo mảng numpy

# Đường thẳng # viết phương trình đường trung trực
y_lin = -(Centroid[0,0] - Centroid[1,0] ) * (x_lin - (Centroid[0,0] + Centroid[1,0])/2) / (Centroid[0,1] - Centroid[1,1] ) + (Centroid[0,1] + Centroid[1,1])/2 # hệ số đường thẳng
plt.scatter(X[ : , 0], X[ : , 1], s =50, c='green',marker = '^')
plt.scatter(Centroid[0,0], Centroid[0,1], s=200, c='b', marker='d')
plt.scatter(Centroid[1,0], Centroid[1,1], s=200, c='r', marker='*')
plt.fill_between(x_lin,y_lin,-2.5,alpha = 0.3)
plt.fill_between(x_lin,y_lin,4.1,alpha = 0.3,color = 'purple')


plt.show()
```

```python
label_train = Kmean.labels_ # Label các điểm dữ liệu
label_train
```

```python
sample_test=np.array([[-3.0,-3.0],[3,2]]) # tạo vecto
second_test=sample_test.reshape(2, -1) # reshape lại thành 1 dimension
Kmean.predict(second_test) # dự đoán label của 2 điểm
```
